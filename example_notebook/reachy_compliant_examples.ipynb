```python
# Importing necessary libraries

from reachy_sdk_modified import ReachySDK
import numpy as np

from reachy_sdk_modified.trajectory import goto_compliant
from reachy_sdk_modified.trajectory import InterpolationMode
import time
```

```python
# Only needed for simulation

import subprocess
```

```python
def happy_antennas():
    for _ in range(7):
        reachy.head.l_antenna.goal_position = 30.0
        reachy.head.r_antenna.goal_position = -30.0

        time.sleep(0.2)

        reachy.head.l_antenna.goal_position = -30.0
        reachy.head.r_antenna.goal_position = 30.0

        time.sleep(0.2)

    reachy.head.l_antenna.goal_position = 0.0
    reachy.head.r_antenna.goal_position = 0.0
```

```python
def hello_left_arm():
    left_arm_hello_position = {
    reachy.l_arm.l_shoulder_pitch: 0,
    reachy.l_arm.l_shoulder_roll: 50,
    reachy.l_arm.l_arm_yaw: 70,
    reachy.l_arm.l_elbow_pitch: -120,
    reachy.l_arm.l_forearm_yaw: 80,
    reachy.l_arm.l_wrist_pitch: 0,
    reachy.l_arm.l_wrist_roll: -40,
    }

    left_arm_base_position = {
    reachy.l_arm.l_shoulder_pitch: 0,
    reachy.l_arm.l_shoulder_roll: 0,
    reachy.l_arm.l_arm_yaw: 0,
    reachy.l_arm.l_elbow_pitch: 0,
    reachy.l_arm.l_forearm_yaw: 0,
    reachy.l_arm.l_wrist_pitch: 0,
    reachy.l_arm.l_wrist_roll: 0,
    }
    
    # The 'goto_compliant' function is exactly like the goto function, except it takes two additional inputs
    
    goto_compliant(
    goal_positions = left_arm_hello_position,
    damping_matrix = np.identity(len(left_arm_hello_position)),
    stiffness_matrix = np.identity(len(left_arm_hello_position)),
    duration = 5,
    interpolation_mode = InterpolationMode.MINIMUM_JERK
    )

    for _ in range(3):
        goto_compliant(
          goal_positions = {reachy.l_arm.l_wrist_roll: 20.0},
          damping_matrix = np.identity(1),
          stiffness_matrix = np.identity(1),
          duration = 1.5,
          interpolation_mode = InterpolationMode.LINEAR
        )
        goto_compliant(
          goal_positions = {reachy.l_arm.l_wrist_roll: -40.0},
          damping_matrix = np.identity(1),
          stiffness_matrix = np.identity(1),
          duration = 1.5,
          interpolation_mode = InterpolationMode.LINEAR
        )

    time.sleep(0.2)

    goto_compliant(
    goal_positions = left_arm_base_position,
    damping_matrix = np.identity(len(left_arm_base_position)),
    stiffness_matrix = np.identity(len(left_arm_base_position)),
    duration = 5,
    interpolation_mode = InterpolationMode.LINEAR
    )
```

```python
def move_head():
    reachy.head.look_at(0.5, -0.3, -0.1, duration=1.0)
    time.sleep(0.3)
    reachy.head.look_at(0.5, 0.3, -0.1, duration=1.0)
    time.sleep(0.3)
    reachy.head.look_at(0.5, 0, -0.1, duration=1.0)
    time.sleep(0.3)
    
    head_tilted_position = {
    reachy.head.neck_roll: -20,
    reachy.head.neck_pitch: 0,
    reachy.head.neck_yaw:0,
    }
    
    goto_compliant(
    damping_matrix = np.identity(len(head_tilted_position)),
    stiffness_matrix = np.identity(len(head_tilted_position)),
    goal_positions = head_tilted_position,
    duration = 1,
    interpolation_mode = InterpolationMode.MINIMUM_JERK
    )
```

```python
def say_hello():
    #move_head()
    happy_antennas()
    hello_left_arm()
    #reachy.head.look_at(0.5, 0, 0, duration=0.5)
```

```python
def get_localhost_wsl():
    result = subprocess.check_output("cat /etc/resolv.conf | grep nameserver | awk '{ print $2 }'", shell=True, text=True)
    result = result[:-1]
    print(result)
    return result
```
